```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
```

```python
sst_ds = xr.open_dataset('./data/sea_surface_temperature.nc')
sst_ds = sst_ds.sel(lats=slice(None, None, -1))
sst_ds
```

```python
sla_ds = xr.open_dataset('./data/sea_level_anomaly.nc')
sla_ds

```

```python

sla_lons = np.array(sla_ds['longitude'])
sla_lats = np.array(sla_ds['latitude'])
sst_lons = np.array(sst_ds['longitude'])
sst_lats = np.array(sst_ds['latitude'])
sla = np.array(sla_ds['sla'])

sla_lons = np.tile(sla_lons,(160,1))
sla_lats = np.tile(sla_lats,(320,1)).transpose()


```

```python
from scipy.interpolate import griddata
resized_sla = []
for t in range(348):

    lons = []
    lats = []
    z_temp = []
    print("{}th time calculating".format(t))
    for i in range(160):
        for j in range(320):
            if sla[t,i,j] == sla[t,i,j]:
                lons.append(sla_lons[i,j])
                lats.append(sla_lats[i,j])
                z_temp.append(sla[t,i,j])



    znew = griddata((lats, lons), z_temp, (sst_lats[None,:], sst_lons[:,None]), method='linear')
    znew = znew.transpose()
    for la in range(40):
        for lo in range(80):
            if sst_ds['sst'][t,la,lo] != sst_ds['sst'][t,la,lo]:
                znew[la,lo] = np.nan

    resized_sla.append(znew)

```
